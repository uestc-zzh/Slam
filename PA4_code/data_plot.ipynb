# Data show

```python
import matplotlib.pyplot as plt
# % matplot inline
import numpy as np
import os
import math
import warnings
warnings.filterwarnings("ignore")
```

## Plot GN

```python
origin_fd = open("origin_data.txt")
out_fd = open("out_data.txt")

origin_data=[]
for data in origin_fd.readlines():
    origin_data.append(float(data))
out_data=[]
for data in out_fd.readlines():
    out_data.append(float(data))
```

```python
plt.figure(figsize=(12,7))
p1=plt.plot(origin_data,color='r',marker='.')
p2=plt.plot(out_data,color='g',marker='.')
plt.title("Gauss-Newton")
plt.grid("true")
plt.xlabel("x")
plt.ylabel("y")
plt.legend([p1,p2],labels=['origin', 'GN'],loc='best',numpoints=1)
plt.show()
```
